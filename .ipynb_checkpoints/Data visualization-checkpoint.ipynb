# Matplotlib:
Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. Matplotlib makes easy things easy and hard things possible.

*Create publication quality plots.
*Make interactive figures that can zoom, pan, update.
*Customize visual style and layout.
*Export to many file formats.
*Embed in JupyterLab and Graphical User Interfaces.
*Use a rich array of third-party packages built on Matplotlib.

```python
import matplotlib.pyplot as plt
list=[28,54,63,56,24]
plt.plot(list)
plt.show()
```

```python
list=[28,54,63,56,24,34,67,88,89,99]
Date=[1,2,4,5,6,9,10,13,25,30]
plt.plot(Date,list, label='amount disbursed',color='r',marker='o',
         markerfacecolor='b')
plt.xlabel("Date")
plt.ylabel("amount")
plt.title("Amount transaction")
plt.legend(loc=1)
plt.show()
```

```python
list=[28,54,63,56,24,34,67,88,89,99]
Date=[1,2,4,5,6,9,10,13,25,30]
plt.plot(Date,list, label='amount disbursed',color='r',marker='o',
         markerfacecolor='b', linestyle='--')
plt.xlabel("Date")
plt.ylabel("amount")
plt.title("Amount transaction")
plt.legend(loc=1)
plt.show()
```

```python
y=[28,54,63,56,24,34,67,88,89,99]
z=[23,24,56,67,47,57,56,69,56,98]
w=[12,34,56,78,90,98,97,96,94,91]
Date=[1,2,4,5,6,9,10,13,25,30]
plt.plot(Date, y, label=' y customer',color='r',marker='o',
         markerfacecolor='b')
plt.plot(Date, z, label=' z customer')
plt.plot(Date, w, label='z customer')
plt.xlabel("Date(august)")
plt.ylabel("amount(lac)")
plt.title("Customer Amount transaction")
plt.legend(loc=1)
plt.show()
```

```python
y=[28,54,63,56,24,34,67,88,89,99]
z=[23,24,56,67,47,57,56,69,56,98]
w=[12,34,56,78,90,98,97,96,94,91]
Date=[1,2,4,5,6,9,10,13,25,30]
plt.plot(Date, y, label=' y customer',color='r',marker='o',
         markerfacecolor='b')
plt.plot(Date, z, label=' z customer')
plt.plot(Date, w, label='w customer')
plt.xlabel("Date(august)")
plt.ylabel("amount(lac)")
plt.xlim(1,20)
plt.ylim(20,150)
plt.title("Customer Amount transaction")
plt.legend(loc=1)
plt.show()
```

```python
y=[28,54,63,56,24,34,67,88,89,99]
z=[23,24,56,67,47,57,56,69,56,98]
w=[12,34,56,78,90,98,97,96,94,91]
Date=[1,2,4,5,6,9,10,13,25,30]
plt.plot(Date, y, label=' y customer',color='r',marker='o',
         markerfacecolor='b')
plt.plot(Date, z, label=' z customer',marker='*',markerfacecolor='r')
plt.plot(Date, w, label='w customer', marker='>',markerfacecolor='y')
plt.xlabel("Date(august)")
plt.ylabel("amount(lac)")
plt.xlim(1,20)
plt.ylim(20,150)
plt.title("Customer Amount transaction")
plt.legend(loc=1)
plt.grid()
plt.show()
```

```python
y=[28,54,63,56,24,34,67,88,89,99]
z=[23,24,56,67,47,57,56,69,56,98]
w=[12,34,56,78,90,98,97,96,94,91]
Date=[1,2,4,5,6,9,10,13,25,30]
plt.plot(Date, y, label=' y customer',color='r',marker='o',
         markerfacecolor='b')
plt.plot(Date, z, label=' z customer')
plt.plot(Date, w, label='w customer')
plt.xlabel("Date(august)")
plt.ylabel("amount(lac)")
plt.xlim(1,20)
plt.ylim(20,150)
plt.title("Customer Amount transaction")
plt.grid(True, linewidth=2,color='b')
plt.legend(loc=1)
plt.show()
```

```python
y=[28,54,63,56,67,34,67,88,89,99]
z=[23,24,56,67,47,57,56,69,56,98]
w=[12,34,56,78,90,98,97,96,94,91]
Date=[1,2,4,5,6,9,10,13,25,30]
plt.plot(Date, y, label=' y customer',color='r',marker='o',
         markerfacecolor='b')
plt.plot(Date, z, label=' z customer',marker='>',markerfacecolor='r')
plt.plot(Date, w, label='w customer',marker='*',markerfacecolor='y')
plt.xlabel("Date(august)")
plt.ylabel("amount(lac)")
plt.xlim(1,20)
plt.ylim(20,150)
plt.title("Customer Amount transaction")
plt.legend(loc=1)
plt.grid(True, linewidth=2, color='b')
plt.show()
plt.savefig('image5', dpi=300, facecolor='r')
```

```python
student=['a','b','c','d','e','f']
marks=[23,45,67,73,67,45]
plt.bar(student, marks, label='student performance', width=0.2, color='g')
plt.xlabel('name of the student')
plt.ylabel('markes of the student')
plt.legend()
```

```python
student=['a','b','c','d','e','f']
marks1=[23,45,67,73,67,45]
marks2=[23,45,56,46,37,78]
plt.bar(student, marks1, label='first sem', width=0.2, color='g')
plt.bar(student, marks2, label='second sem', width=0.2, color='r')
plt.xlabel('name of the student')
plt.ylabel('markes of the student')
plt.legend()
```

```python
student=['a','b','c','d','e','f']
marks1=[23,45,67,73,67,45]
marks2=[23,45,56,46,37,78]

xps=np.arange(len(student))
plt.bar(xps+0.4, marks1, label='first sem', width=0.2, color='g')
plt.bar(student, marks2, label='second sem', width=0.2, color='r')
plt.xlabel('name of the student')
plt.ylabel('markes of the student')
plt.legend()
```

```python
#piechart

ev=[40,23,6,23,56]
en=['reent','other','phone','food','car']
plt.pie(ev, labels=en)
plt.show()
```

```python
plt.axis('equal')
plt.pie(ev,labels=en, shadow=True,autopct='%1.1f%%',radius=1.2)
plt.show()
```

Statistical measures:
    
    1. Measures of frequency:
        numbers of occurence
    2. Measures of central tendency:
        weather the data values accumilate in the middle of distribution 
        
         1. Mean.
         2. Meadian: middle value of a ordered list
         3. Mode: Mode is the number that repeated more often   
            
    3. Measures of spread:
          1. standard deviation 
            2. variance
            3. Quartile
        
    4. Measures of position: Position identify the exact location 
        of a particular data
